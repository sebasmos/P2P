````python
import pickle
with open('loss.pkl', 'rb') as f:
    Loss_per_epoch = pickle.load(f)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.DataFrame(Loss_per_epoch)
columns = {0:'Hoechst',
           1:'AF1',
           2:'CD31',
           3:'CD45',
           4:'CD68',
           5:'Blank',
           6:'CD4',
           7:'FOXP3',
           8:'CD8a',
           9:'CD45RO',
           10:'CD20',
           11:'PD-L1',    
           12:'CD3e',
           13:'CD163',
           14:'E-Cadherin',
           15:'PD-1',
           16:'Ki-67',
           17:'Pan-CK',
           18:'SMA'
            }
df.rename(columns=columns, inplace=True)
df
````

````python
n_ch= len(df.columns)
````

````python
df.iloc[:,:n_ch-1]
````

````python
import seaborn as sns
palette = sns.color_palette("husl", (n_ch-1)) # https://seaborn.pydata.org/generated/seaborn.color_palette.html
fig = plt.figure('fig_size', (10, 10))
# Adding grid
sns.set(style="ticks")
#sns.set_style("darkgrid")#whitegrid")# #https://seaborn.pydata.org/generated/seaborn.set_style.html
sns.lineplot(data=df.iloc[:,:n_ch-1], 
             markers=True, 
             dashes=False, 
             palette=palette, 
             linewidth = 5)

plt.legend(loc='lower right')
plt.title("L1 per channel - L1{Fake_IF, Real_IF}")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.grid()

#plt.legend(labels=['legendEntry1', 'legendEntry2', 'legendEntry3', 'legendEntry2', 'legendEntry3','legendEntry2', 'legendEntry3','legendEntry2', 'legendEntry3','legendEntry2', 'legendEntry3','legendEntry2', 'legendEntry3','legendEntry2', 'legendEntry3','legendEntry2', 'legendEntry3','legendEntry2', 'legendEntry3'])

plt.show()
plt.close('all')
````

````python
import seaborn as sns
palette = sns.color_palette("husl", n_ch-1) # https://seaborn.pydata.org/generated/seaborn.color_palette.html
fig = plt.figure('fig_size', (10, 10))
# Adding grid
sns.set(style="ticks")
sns.set_style("darkgrid")#whitegrid")# #https://seaborn.pydata.org/generated/seaborn.set_style.html
sns.lineplot(data=df.iloc[:,:n_ch-1], 
             markers=True, 
             dashes=False, 
             palette=palette, 
             linewidth = 5)

plt.title("L1 per channel - L1{Fake_IF, Real_IF}")
plt.axhline(y=17.5 , color='red', linestyle='--', label="Limit")
plt.legend(loc='lower right')
plt.ylabel("Loss")
plt.xlabel("Epochs")
#plt.grid()
plt.savefig("loss1.png", bbox_inches='tight', pad_inches=0)#, dpi = 64)
plt.show()
plt.close('all')
````
